```python
# Testing Advanced Imputer with a Different Dataset (Diabetes)
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from tabularaml.preprocessing.pipeline import PipelineWrapper
from tabularaml.preprocessing.imputers import AdvancedImputer, SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Load the diabetes dataset
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = pd.Series(diabetes.target)

# Create a more challenging missing data pattern (30% missing, not at random)
# Make more missingness in higher values to create a bias
X_missing = X.copy()
for col in X.columns:
    # Create a biased missing pattern: higher values are more likely to be missing
    threshold = X[col].quantile(0.7)  # Higher 30% of values
    higher_values = X[col] > threshold
    # 80% of high values will be missing, 10% of lower values will be missing
    high_mask = (higher_values) & (np.random.rand(len(X)) < 0.8)
    low_mask = (~higher_values) & (np.random.rand(len(X)) < 0.1)
    missing_mask = high_mask | low_mask
    X_missing.loc[missing_mask, col] = np.nan

print(f"Dataset shape: {X_missing.shape}")
print(f"Missing values count: {X_missing.isna().sum().sum()}")
print(f"Percentage of missing values: {X_missing.isna().sum().sum() / (X_missing.shape[0] * X_missing.shape[1]) * 100:.2f}%")

# Display missing pattern by column
missing_by_col = X_missing.isna().sum().sort_values(ascending=False)
print("\nMissing values by column:")
for col, count in missing_by_col.items():
    print(f"{col}: {count} missing ({count/len(X_missing)*100:.1f}%)")

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_missing, y, test_size=0.2, random_state=42)

# Use a more powerful model that can handle complex relationships
model = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Baseline model with simple imputation (mean)
baseline_pipeline = Pipeline([
    ('imputer', SimpleImputer('mean')),
    ('scaler', StandardScaler()),
    ('model', model)
])

# Advanced imputer model using the wrapper
wrapper = PipelineWrapper(imputer=AdvancedImputer())
advanced_pipeline = wrapper.get_pipeline(X_train, y_train)
advanced_pipeline.steps.append(('model', model))

# Cross-validation with 5 folds
print("\nBaseline Model Performance (Mean Imputation):")
baseline_scores = cross_val_score(baseline_pipeline, X_train, y_train, 
                                  cv=5, scoring='r2')
print(f"Cross-validation R² scores: {baseline_scores}")
print(f"Mean R² score: {baseline_scores.mean():.4f} ± {baseline_scores.std():.4f}")

print("\nAdvanced Imputer Model Performance:")
advanced_scores = cross_val_score(advanced_pipeline, X_train, y_train, 
                                 cv=5, scoring='r2')
print(f"Cross-validation R² scores: {advanced_scores}")
print(f"Mean R² score: {advanced_scores.mean():.4f} ± {advanced_scores.std():.4f}")

# Final evaluation on test set
baseline_pipeline.fit(X_train, y_train)
advanced_pipeline.fit(X_train, y_train)

baseline_pred = baseline_pipeline.predict(X_test)
advanced_pred = advanced_pipeline.predict(X_test)

print("\nTest Set Results:")
print(f"Baseline RMSE: {np.sqrt(mean_squared_error(y_test, baseline_pred)):.4f}")
print(f"Advanced RMSE: {np.sqrt(mean_squared_error(y_test, advanced_pred)):.4f}")
print(f"Baseline R²: {r2_score(y_test, baseline_pred):.4f}")
print(f"Advanced R²: {r2_score(y_test, advanced_pred):.4f}")

# Visualize results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Plot 1: Feature importance if available
plt.subplot(1, 2, 1)
try:
    feature_importances = baseline_pipeline.named_steps['model'].feature_importances_
    sorted_idx = np.argsort(feature_importances)
    plt.barh(np.array(diabetes.feature_names)[sorted_idx], feature_importances[sorted_idx])
    plt.title('Feature Importance')
except:
    plt.text(0.5, 0.5, "Feature importance not available", ha='center', va='center')
    plt.title('Feature Importance Not Available')

# Plot 2: Prediction comparison
plt.subplot(1, 2, 2)
plt.scatter(y_test, baseline_pred, alpha=0.5, label='Baseline (Mean Imputation)')
plt.scatter(y_test, advanced_pred, alpha=0.5, label='Advanced Imputer')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Prediction Comparison')
plt.legend()

plt.tight_layout()
plt.show()
```
